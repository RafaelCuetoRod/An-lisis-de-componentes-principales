# **PCA utilizando la biblioteca Sklearn**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# **Visualización de un conjunto de datos de alta dimensión con PCA usando Sklearn**

# Normalización del conjunto de datos

```python
df_st =  StandardScaler().fit_transform(data0)  
pd.DataFrame(df_st, columns=data0.columns).head()
```

# **Estandarización**

# Perform PCA using scikit-learn!

```python
pca_out = PCA(n_components=10).fit(df_st)

pca_out.explained_variance_ratio_
```

Se realiza descomposición por componentes principales en este caso 10 y se ordenan de acuerdo a su importancia

```python
# Cumulative proportion of variance   
np.cumsum(pca_out.explained_variance_ratio_)
```

Se realiza la suma acumulada de los componentes principales, en este caso con 5 componentes principales ya tenemos un 78,59% de la variación del conjunto de datos

```python
loadings = pca_out.components_
num_pc = pca_out.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = data0.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

# **Matiz de correlación**

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,12))
ax = sns.heatmap(loadings_df, annot=True, cmap='bwr')
plt.show()
```

Segun la matriz de correlación podemos decir que:

- Las variables temp, DC, DMC, FFMC tienen una alta correlación negativa con la componente principal 1
- Las variables X, Y tienen una alta correlación positiva con la componente principal 2
- La variable RH tiene una alta correlación positiva con la componente principal 3
- La variable wind tiene una alta correlación positiva con la componente principal 4
- La variable rain tiene una alta correlación positiva con la componente principal 5
- La variable wind tiene una alta correlación negativa con la componente principal 6
- Las variables X, FFMC tienen una alta correlación negativa con la componente principal 7
- Las variables Y, FFMC tienen una alta correlación negativa con la componente principal 8
- Con la componente principal 9, la variable DMC tiene una alta correlación negativa y la variable DC tienen una alta correlación positiva
- Las variables temp, RH tienen una alta correlación negativa con la componente principal 10

# **Retención de los componentes principales (PC)**

```python
plt.figure(figsize=(12,5))
plt.bar(loadings_df.columns,pca_out.explained_variance_ratio_*100, color="k")
plt.ylabel("Proportion of variance (%)")
plt.xlabel("PCs")
plt.show()
```

A traves del grafico se puede observar que con las primeros cinco (5) componetes principales se tiene un porcentaje aproximado del 79% de la variación del conjunto de datos

```python
plt.figure(figsize=(12,5))
plt.bar(loadings_df.columns,np.cumsum(pca_out.explained_variance_ratio_*100))
plt.ylabel("Cumulative proportion of variance (%)")
plt.xlabel("PCs")
plt.show()
```

Segun el grafico las variables temp, DC, FFMC, DMC, ISI representan una alta correlación con el componente principal 1; y las variables X, Y representan una alta correlación con el componente principal 2.

# **Biplot del PCA**

```python
cmap=plt.cm.seismic_r
norm=plt.Normalize(vmin=0, vmax=1090.84)
```

```python
# get PC scores
pca_scores = PCA().fit_transform(df_st)

# get 2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=data0.columns.values,dotsize=16,
               colordot=cmap(norm(data["area"])),arrowcolor='r',arrowlinewidth=0.9,arrowlinestyle='--',show=True,dim=(12,10),axlabelfontsize=14,
               var1=round(pca_out.explained_variance_ratio_[0]*100, 2),
               var2=round(pca_out.explained_variance_ratio_[1]*100, 2))
```

# **Interpretación del ACP**

Los cinco (5) primeros componentes principales contribuyen con ~79% de la variación total del conjunto de datos y tienen valores propios > 1 y por lo tanto proporciona una buena aproximación de la variación presente en el conjunto de datos original. El límite de la variación acumulada del 70% es común para retener las PC para el análisis.

A partir de la gráfica de biplot y cargas, podemos ver que las variables temp, DC, DMC, FFMC, ISI están altamente asociadas y forman un grupo. de forma similiar las variables X, Y estan altamente asociados y forman otro grupo.

La longitud de las PC en biplot se refiere a la cantidad de varianza aportada por las PC. Cuanto mayor sea la longitud de la PC, mayor será la varianza aportada y bien representada en el espacio.
